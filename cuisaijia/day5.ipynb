Matplotlib画图

```python
#折线图
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

x = np.arange(9)
y = np.sin(x)
z = np.cos(x)
# marker数据点样式，linewidth线宽，linestyle线型样式，color颜色
plt.plot(x,y,marker = '*', linewidth = 3, linestyle = '--', color = 'orange')
plt.plot(x,z)
plt.title("matplotlib example")
plt.xlabel("height")
plt.ylabel("width")
#设置图例
plt.legend(["Y","Z"], loc = "upper right")
plt.grid(True)
plt.show()
```

```python
#散点图
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x,y)
plt.show()
```

```python
#柱状图
x = np.arange(10)
y = np.random.randint(0,30,10)
plt.bar(x,y)
plt.show()
```

```python
#饼图
x = np.random.randint(1,10,3)
plt.pie(x)
plt.show()
```

```python
#直方图
mean,sigma = 0,1
x = mean + sigma * np.random.randn(10000)
plt.hist(x,50)
plt.show()
```

```python
#箱线图
fig ,axes = plt.subplots(2,1,figsize = (10,6))#figsize图像标签大小
df = pd.DataFrame(np.random.rand(10,5),columns = ['a','b','c','d','e'])#生成10*5的（0，1）之间的随机数，分别对应columns参数 A B C D 所在的位置；
# whiskers：分位数与error bar横线之间竖线的颜色 medians：中位数线颜色 caps：error bar横线颜色
color = dict(boxes = 'DarkGreen',whiskers = 'DarkOrange',medians = 'DarkBlue')

# vert:是否垂直，默认为True ; position:箱线图占位;也可以用df.boxplot()
df.plot.box(ylim = [0,1.2], grid = True, color = color, ax = axes[0])
df.plot.box(vert = False, positions = [1,4,5,6,8], color = color, ax = axes[1])
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
        
with sns.axes_style('darkgrid'):
    plt.subplot(2,1,1)
    sinplot()
plt.subplot(2,1,2)
sinplot(-1)
```
