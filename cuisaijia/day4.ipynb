```python
import pandas as pd
import numpy as np
#from datetime import datetime as dt

df = pd.DataFrame(pd.read_csv('C:/Users/lenovo/Desktop/csj/hancheng.csv'))

#替换空值
#print(df.isnull().any())
#print(df[df.isnull().values == True])
df = df.fillna(-99)
df = df.replace('缺失',-99)

#datetime
df['生日'] = pd.to_datetime(df['生日'],format = '%Y.%m.%d') #将生日转换为标准格式
df['生日'] = df['生日'].dt.strftime('%Y-%m-%d') #去掉日期中的时、分、秒


date = df['生日'].str.split('-',3,True) 
date.columns = ['年','月','日'] #将日期分成三列
df = pd.concat([df,date],axis = 1) #横向拼接
df= df.drop('生日',axis = 1) #删除原来的生日属性列
#pd.to_datetime.dt.year('生日')


#df.apply(pd.to_numerics,errors = 'ignore')#数据类型转换
df['年'] = pd.DataFrame(df['年'],dtype = int)
df['月'] = pd.DataFrame(df['月'],dtype = int)
df['日'] = pd.DataFrame(df['日'],dtype = int)
df['年龄'] = pd.DataFrame(df['年龄'],dtype = int) 
df['裸右'] = pd.DataFrame(df['裸右'],dtype = float)
df['裸左'] = pd.DataFrame(df['裸左'],dtype = float)



strange_rows = df['这一个戴镜时间'].str.contains('>').fillna(False)
df = df.replace(strange_rows,1)
#print(df)
df.describe()
#print(df.dtypes())
```
