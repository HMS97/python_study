```python
#缺失值改为-99
import pandas as pd
import numpy as np
import os
os.getcwd()
df = pd.DataFrame(pd.read_csv('hancheng.csv'))
a=df.fillna({'生日':-99,'年龄':-99,'裸右':-99,'裸左':-99,'带镜右眼':-99,'带镜左眼':-99,'电脑验光右':-99,'电脑验光左':-99,'这一个戴镜时间':-99,'总戴镜时间':-99})
print(a)
```

```python
#使用describe函数
c = a.describe()
print(c)
```

```python
a['生日'] = pd.to_datetime(a['生日'],format = '%Y.%m.%d') #将生日转换为标准格式
a['生日'] = a['生日'].a.dt.strftime('%Y-%m-%d') #去掉日期中的时、分、秒
date = a['生日'].str.split('-',3,True) 
date.columns = ['年','月','日'] #将日期分成三列
a = pd.concat([a,date],axis = 1) #横向拼接
a= pd.drop('生日',axis = 1) #删除原来的生日属性列
```

```python
#修改数据类型
a['年'] = pd.DataFrame(a['年'],dtype = int)
a['月'] = pd.DataFrame(a['月'],dtype = int)
a['日'] = pd.DataFrame(a['日'],dtype = int)
a['年龄'] = pd.DataFrame(a['年龄'],dtype = int) 
a['裸右'] = pd.DataFrame(a['裸右'],dtype = float)
a['裸左'] = pd.DataFrame(a['裸左'],dtype = float)
```

```python
#将不合理符号修改
strange_rows = a['这一个戴镜时间'].str.contains('>').fillna(False)
a = a.replace(strange_rows,1)
```
