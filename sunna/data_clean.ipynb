```python
import pandas as pd
# 读取csv文件
df = pd.read_csv(('C:\\Users\\dadada\\Desktop\\sunflower\\python_study\\homework\\hancheng.csv'))
```

```python
df =  df[df.columns[:13]]
```

```python
df = df.fillna(-99)
```

```python
df = df.replace('缺失', -99)
df = df.replace(' ', -99)
df = df.replace('  ', -99)
```

```python
## borthday convert to datetime
df['生日'] = pd.to_datetime(df['生日'])
```

```python
df['年'] = df['生日'].dt.year
df['月'] = df['生日'].dt.month
df['日'] = df['生日'].dt.day
```

```python
df['性别'] = df['性别'].map(lambda x: 1 if x == '男' else 0)
```

```python
df['带镜左眼'].unique()
```

```python
df['带镜右眼'] = df['带镜右眼'].astype('float')
```

```python
df['带镜左眼'] = df['带镜左眼'].astype('float')
```

```python
df['电脑验光右'] = pd.to_numeric(df['电脑验光右'], errors='coerce')

```

```python
df['电脑验光左'] = pd.to_numeric(df['电脑验光左'], errors='coerce')
```

```python
for i in df.columns[-5:-3]:
    df[i] = pd.to_numeric(i, errors='coerce')
```

```python
df['年龄'] = pd.to_numeric(df['年龄'], errors='coerce')
```

```python
df = df.drop(df[df['年龄']>100].index, axis = 0 )



```

```python

df.to_csv('result.csv', index = False)
```
