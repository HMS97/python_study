```python
sum = 0
i = 1
while i>0:
 if i>100:
  break
 else:
  sum += i
  i = i+1
print('1加到100的和为',sum)
```

```python
list1 = [x for x in range(200) if x%2==0]
print('0到200之间的偶数:',list1)
list2 = [x for x in range(200) if x%3==0]
print('0到200之间可被3整除的数:',list2)
a = set(list1)
b = set(list2)
c = a&b
print('与集',c)
d = a|b
print('或集',d)
e = a-b
print('差集',e)
```
