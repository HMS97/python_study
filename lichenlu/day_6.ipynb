# 了解get和post请求

```python
import requests
r = requests.get('https://api.github.com/events')
```

```python
r = requests.post('http://httpbin.org/post', data = {'key':'value'})
```

```python
#传递字符到httpbin.org/get
payload = {'key1': 'value1', 'key2': 'value2'}
r = requests.get("http://httpbin.org/get", params=payload)
#打印该输出URL
print(r.url)
```
