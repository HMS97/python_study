```python
import numpy as np #导入numpy库
import pandas as pd #导入pandas库
from datetime import datetime as dt

df = pd.DataFrame(pd.read_csv('hancheng.csv')) #读取hancheng.csv文件
df = df.dropna(axis = 1) #删除全部为空的列值


df_1 = df.fillna(-99) #用-99替换NaN值
df_2 = df_1.replace('缺失',-99) #用-99替换缺失值

 
df_2['生日'] = pd.to_datetime(df_2['生日'],format = '%Y.%m.%d') #将生日转换为标准格式
df_2['生日'] = df_2['生日'].dt.strftime('%Y-%m-%d') #去掉日期中的时、分、秒


date = df_2['生日'].str.split('-',3,True) 
date.columns = ['年','月','日'] #将日期分成三列
df_3 = pd.concat([df_2,date],axis = 1) #横向拼接
df_3 = df_3.drop('生日',axis = 1) #删除原来的生日属性列


df_3['年龄'] = pd.DataFrame(df_3['年龄'],dtype = int) #将年龄-int、裸右-float、裸左、float
df_3['裸右'] = pd.DataFrame(df_3['裸右'],dtype = float)
df_3['裸左'] = pd.DataFrame(df_3['裸左'],dtype = float)
df_3['年'] = pd.DataFrame(df_3['年'],dtype = int)
df_3['月'] = pd.DataFrame(df_3['月'],dtype = int)
df_3['日'] = pd.DataFrame(df_3['日'],dtype = int)


now = pd.Timestamp.now() #获取当前的年份
df_3['年龄'] = now.year - df_3['年'] #计算并更正年龄

#获取 这一个戴镜时间 数据列中含有<的数据
rows_with_qiguai = df_3['这一个戴镜时间'].str.contains('＜').fillna(False)
df_3[rows_with_qiguai]
#将含有<的数据转换为0
df_4 = df_3.replace(rows_with_qiguai,0)


```
