```python
def add_1():
    i = 1 
    add_1 = 0
    while i < 101:
        add_1 = add_1 + i
        i = i+1
    return add_1
print(add_1())

def add_2():
    add_2 = 0
    i=1
    while i<101:
        add_2 = add_2 + i
        i = i+1
        if i == 50:
            add_2 = add_2 + 0           
            i = i+1	
            continue
    return add_2
print (add_2())
```

```python
###########4_1订正#####################
list_1 = []
list_2 = []
for a in range(201):
    if a%2 == 0:
       list_1.append(a) 
for b in range(201):
    if b%3 == 0:
        list_2.append(b)
print('0~200之间被2整除的数:',list_1)
print('0~200之间被3整除的数:',list_2)
m = set(list_1)
n = set(list_2)
print('与集:',m & n)
print('或集:',m | n)
print('差集:',m - n)

```
