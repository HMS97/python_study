# Maplotlib画图

```python
#画折线图
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-1, 2, .01)
y = np.sin(2 * np.pi * x)
#marker数据点样式、linewidth线宽，linestyle线条样式，color颜色
plt.plot(x, y,marker = "*",linewidth = 2,linestyle = '--',color = 'green')
plt.plot(x, y)
plt.title("Maplotlib_zhexiantu")
plt.grid(True)
plt.show()
```

```python
#画散点图
x = np.random.normal(0, 1, 20)
y = np.random.normal(0, 1, 20)
t = np.arctan2(y, x)
#scatter函数存放下x，y数据，s表示圆点面积，c表示颜色，alpha表示透明度
plt.scatter(x, y,s=75, c = t, alpha = 0.7)
#设置x、y轴范围
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
```

```python
#画柱状图
x = np.arange(50)
y = np.random.randint(0, 30, 50)#randint(low, high, size),生成low到high之间（半开区间{low,high)）,size个数据
plt.bar(x, y)
plt.show()
```

```python
#饼图
labes=['A','B','C','D']
fracs=[15,30,45,10]
#扇形四个"邻居"之间的缝隙
explode=[0,0.1,0,0]
#设置x,y轴比例为1：1，从而达到一个正的圆
plt.axes(aspect=1)
#labels标签参数,x是对应的数据列表,autopct显示每一个区域占的比例,explode突出显示某一块,shadow=True表示有一定的阴影看起来更立体
plt.pie(x=fracs,labels=labes,autopct="%.0f%%",explode=explode,shadow=True)
plt.show()
```

```python
#画箱线图
#normal(loc=0.0, scale=1.0, size=None)
#loc：float 此概率分布的均值（对应着整个分布的中心centre）
#scale：float 此概率分布的标准差（对应于分布的宽度，scale越大越矮胖，scale越小，越瘦高）
#size：int or tuple of ints 输出的shape，默认为None，只输出一个值

data = np.random.normal(loc = 0, scale = 1, size = 1000)
#sys 点的形状，whis虚线的长度
plt.boxplot(data, sym = '*', whis = 1.5)
plt.show()
```

```python
#seaborn进阶

import seaborn as sns
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
# 绘制小提琴图
ax = sns.violinplot(x=tips["total_bill"])
```

```python
#绘制偏移正弦波
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
#切换到seaborn模式
sns.set_style("ticks")
sns.despine() #移除顶部和右边的坐标轴
sns.set_context("paper") #按相对尺寸的顺序（线条越来越粗），分别是paper、notebook、talk、poster
sinplot()
```
